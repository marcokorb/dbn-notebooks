```python
# This notebook source link is: http://www-desir.lip6.fr/~phw/aGrUM/docs/last/notebooks/01-tutorial.ipynb.html

%matplotlib inline
from pylab import *
import matplotlib.pyplot as plt
import os
```

```python
import pyAgrum as gum
```

```python
# Creates the BN model
bn = gum.BayesNet('Pitagoras')
print(bn)
```

```python
# Create nodes

# Subjects nodes
subjects_nodes = ['ca', 'co', 'h']
ca, co, h = [bn.add(name, 2) for name in subjects_nodes]

print('Subjects Nodes')
print(ca, co, h)

# Evidences nodes
evidences_nodes = ['evca', 'evco', 'evh']
evca, evco, evh = [bn.add(name, 2) for name in evidences_nodes]

print('Evidences Nodes')
print(evca, evco, evh)

print(bn)
```

```python
# Create arcs

# Subjects arcs
bn.addArc(ca, h)
bn.addArc(co, h)

# Evidences arcs
bn.addArc(ca, ev_ca)
bn.addArc(co, ev_co)
bn.addArc(h, ev_h)

print(bn)
```

```python
# Graphic visualization
import pyAgrum.lib.notebook as gnb

gnb.forDarkTheme()

bn
```

```python
# Create CPTs

# Method orders for probabilities
# .fillWith([False, True])

# Subjects CPTs
bn.cpt(ca).fillWith([0.468176, 0.531824])
bn.cpt(co).fillWith([0.468176, 0.531824])
bn.cpt(h)[{'ca': 0, 'co': 0}] = [0.908971, 0.091029]
bn.cpt(h)[{'ca': 0, 'co': 1}] = [0.650269, 0.349731]
bn.cpt(h)[{'ca': 1, 'co': 0}] = [0.650269, 0.349731]
bn.cpt(h)[{'ca': 1, 'co': 1}] = [0.124852, 0.875148]


# Evidences CTPs
bn.cpt(evca)[{'ca': 0}] = [0.8, 0.2]
bn.cpt(evca)[{'ca': 1}] = [0.1, 0.9]

bn.cpt(evco)[{'co': 0}] = [0.8, 0.2]
bn.cpt(evco)[{'co': 1}] = [0.1, 0.9]

bn.cpt(evh)[{'h': 0}] = [0.8, 0.2]
bn.cpt(evh)[{'h': 1}] = [0.1, 0.9]
```

```python
# Print ca CPT
bn.cpt(ca)
```

```python
# Print h CPT
bn.cpt(h)
```

```python
gnb.forDarkTheme()

gnb.sideBySide(bn,gnb.getInference(bn))
```

```python
# Inference in BN

ie = gum.LazyPropagation(bn)
```

```python
ie.makeInference()
print (ie.posterior(h))
```

```python
ie.posterior(h)
```

```python
ie.setEvidence({'evca': 1})
ie.makeInference()
ie.posterior(ca)
```

```python
# Soft evidence

ie.setEvidence({'evca': [0.5, 1]})
ie.makeInference()
ie.posterior(ca) # using gnb's feature
```
